```python
import requests
import random
```

```python
path = "/api/events"
base_url = "http://localhost:8002"
create_endpoint = f"{base_url}{path}"

event_count = 10_000
pages = ['/about', '/contact', '/pages', '/pricing']

for i in range(event_count):
    page = random.choice(pages)
    response = requests.post(create_endpoint, json={"page": page})
    if response.ok:
        data = response.json()
        print(data, type(data))
    else:
        print(response.text)
```

```python
detail_path = "/api/events/10000"
detail_endpoint = f"{base_url}{detail_path}"

detail_response = requests.get(detail_endpoint)
print(detail_response.status_code, detail_response.text)

if detail_response.ok:
    data = detail_response.json()
    print(data)
```
